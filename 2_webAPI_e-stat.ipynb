# e-StatのWeb API

```python
import requests
import pandas as pd
import json
import time
import io
```

## 統計表情報の取得

```python
# 別ファイルに保存しておいたAPI tokenを読み込む
with open('e-stat_api_key.json') as f:
    api_key = json.load(f)
```

```python
# 今回の統計表の検索に使うキーワード
searchYear='2020'
searchWord='観光'
```

```python
# 統計表情報取得のためのEndpoint
# api_keyをappIDの値として指定
r=requests.get('http://api.e-stat.go.jp/rest/3.0/app/json/getStatsList?appId={}&searchWord={}&searchYear={}'
               .format(api_key['appId'],searchWord,searchYear))
```

```python
stats_list=r.json()
stats_list
```

```python
stats_list['GET_STATS_LIST'].keys()
```

```python
stats_list['GET_STATS_LIST']['DATALIST_INF']['TABLE_INF']
```

```python
df_stats_list=pd.json_normalize(stats_list['GET_STATS_LIST']['DATALIST_INF']['TABLE_INF'])
```

```python
len(df_stats_list) # pandasの行数を確認
```

## 統計表からの統計データの取得

```python
statsDataId="0003289292" #観光地域経済調査（確報）全国集計表のID
```

```python
# 実際の統計情報取得のためのEndpoint
time.sleep(1)
r=requests.get('http://api.e-stat.go.jp/rest/3.0/app/getSimpleStatsData?appId={}&statsDataId={}&sectionHeaderFlg={}&area'
               .format(api_key['appId'],statsDataId,'2'))
```

```python
r.text
```

```python
df=pd.read_csv(io.StringIO(r.text))
df
```

```python
df.to_csv('{}.csv'.format(statsDataId),encoding='utf_8_sig') # csvとして保存
```

ループで回せば、searchYear='2020'searchWord='観光'に対応するデータが全てダウンロードできる！！！
