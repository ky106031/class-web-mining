# ホットペッパーグルメのAPI

```python
import requests
import pandas as pd
import json
import time
import io
```

## 統計表情報の取得

```python
# 別ファイルに保存しておいたAPI tokenを読み込む
with open('hotpepper_api_key.json') as f:
    api_key = json.load(f)
```

```python
# ホットペッパーグルメのEndpoint
# APIキーをkeyの値として指定

# 滋賀大学の緯度と経度
lat = 35.2688229
lng = 136.222625
# 指定座標からの距離 1が300mに対応。最大5
range = 4 

r=requests.get('http://webservice.recruit.co.jp/hotpepper/gourmet/v1/?key={}&format={}&lat={}&lng={}&range={}'
               .format(api_key['appId'],"json", lat,lng,range))
```

```python
dic = r.json() # Json文字列からPythonのdict形式に変換
dic
```

```python
# これをpandasに変換
df=pd.json_normalize(dic['results']['shop']) 
pd.set_option('display.max_columns', None)
df
```
